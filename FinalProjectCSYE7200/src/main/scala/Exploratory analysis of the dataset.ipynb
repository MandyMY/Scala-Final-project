```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df_a = pd.read_csv('../../../data/Accidents.csv')
```

```python
df_a.head()
```

```python
df_a.describe()
```

```python
sns.heatmap(df_a.isnull(), cbar=False)
```

```python
plt.figure(figsize=(16,12))
sns.heatmap(data=df_a.iloc[:,2:].corr(),annot=True,fmt='.2f',cmap='coolwarm')
plt.show()
```

```python
a = df_a[["Accident_Severity","Number_of_Vehicles", "Number_of_Casualties", "Road_Type",
          "Junction_Detail", "Pedestrian_Crossing-Physical_Facilities", 
          "Light_Conditions", "Weather_Conditions",  
          "Carriageway_Hazards", "Did_Police_Officer_Attend_Scene_of_Accident"]]
```

```python
sns.pairplot(a)
plt.show()
```
